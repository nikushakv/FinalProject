```python
import pandas as pd
cesko_data = pd.read_csv('Statistics of CESKO.csv')
```

```python
print(cesko_data.describe())
```

```python
import seaborn as sns
sns.histplot(cesko_data['Perc Turnout'], bins=10)
```

```python
import matplotlib.pyplot as plt
plt.scatter(cesko_data['Perc Turnout'], cesko_data['GD Share of the turnout'], alpha=0.6)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation analysis
numeric_cols = cesko_data.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
correlation_matrix = numeric_cols.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
